```
#RIFA BOLETOS Cuadrilla 68
import random
import os

#Creamos un array para el almacenamiento de los datos
listaganadores = []

#Declaramos nuestras variables
nganadores = int(input("Ingresa el número de ganadores a seleccionar: "))
nsemana = int(input("Ingresa el número de semana: "))
liminferior = int(input("Ingrese el límite inferior: "))
limsuperior = int(input("Ingrese el límite superior: "))

#Usamos la librería para determinar los ganadores de manera completamente aleatoria y los agregamos a nuestro arreglo
listaganadores.append(random.sample(range(liminferior, limsuperior), nganadores))

#Creamos nuestro resultado agregandole un texto descriptivo
ganadoresl = ("Los ganadores de esta semana", nsemana, "son:", listaganadores)

#Creamos un archivo para almacenar a los ganadores y almacenamos los resultados
archivo = open('/content/gdrive/My Drive/resultadosrifa/ganadores.txt', 'a')
archivo.write(str(ganadoresl))
archivo.close

#Imprimimos la lista de ganadores publicadas en el archivo txt
archivo = open("/content/gdrive/My Drive/ResultadosRifa/ganadores.txt", "r")
print(archivo.read())

#Evitamos que se cierre la consola en ejecución
os.system("pause")
```
